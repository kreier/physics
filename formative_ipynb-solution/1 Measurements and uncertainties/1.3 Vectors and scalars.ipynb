# 4. A car travels at 80. km h -1 up a ramp making an angle of $20^o$with the horizontal. Find its horizontal component. Include a sketch.

Given
- The car velocity: $80kmh^{-1}$
- The angle the car travels: $20^o$

Formula
- Vhorizontal=V*cos$\theta$

Solution
- The sketch
![alt text](https://drive.google.com/uc?id=1zZagAVnHFZ-a0U7HA7Ndiub3XQbEij7s)

```
#solution
import math
v=80          #The car speed up the ramp
angle=20      #The angle of the ramp

vh= v*math.cos(angle)
print('The horizontal component is ' +str(vh)+' kmh-1')
```
